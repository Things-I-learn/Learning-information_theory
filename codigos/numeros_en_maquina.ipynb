```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the floating-point system
beta = 2  # base
t = 3     # number of significant digits
L = -2    # minimum exponent
U = 3     # maximum exponent

# Generate all positive floating-point numbers
x = []
for i in np.arange(1, 2, 0.25):
    for j in range(L, U + 1):
        x.append(i * beta**j)

# Add negative numbers and zero
x = np.array(x)
x = np.concatenate((-x, [0], x))

# Sort numbers
x = np.sort(x)

# Plot the floating-point numbers
y = np.zeros(len(x))
plt.plot(x, y, 'b+')
plt.ylim(-1, 1)
plt.xlabel("Floating-point values")
plt.title("Floating-point Representation")
plt.show()
```
