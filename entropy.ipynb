```python
import numpy as np
import matplotlib.pyplot as plt

def shannon_entropy(p):
    if p == 0 or p == 1:
        return 0
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Generamos valores de probabilidad entre 0 y 1
p_values = np.linspace(0, 1, 1000)
entropy_values = [shannon_entropy(p) for p in p_values]

# Graficamos la entropa
plt.figure(figsize=(10, 6))
plt.plot(p_values, entropy_values, label='Entropa de Shannon')
plt.xlabel('Probabilidad p')
plt.ylabel('Entropa H(X)')
plt.title('Entropa de Shannon para un sistema binario')
plt.grid(True)
plt.legend()
plt.show()
```

- No hay azar cuando un evento tiene toda la probabilidad de ocurrir o cuando no tiene nada 
de probabilidad de ocurrir.

- La entropia maxima es cuando el evento esta entre no ocurrir o ocurrir.

- Si nos movemos un poco en la probabilidad no hay un cambio brusco en nuestra entropia.
